```python
import pickle
import lightgbm as lgb
import pandas as pd
from sklearn.utils import shuffle
import numpy as np
from sklearn import preprocessing

with open('./data/train-id4-crowd-grid5.txt', 'rb') as data_file:
    train = pickle.load(data_file)
with open('./data/test-id4-crowd-grid5.txt', 'rb') as data_file:
    test = pickle.load(data_file)
    
train['Source_Station_encode'] = train['Source_Station_encode'].astype('str') 
train['Target_Station_encode'] = train['Target_Station_encode'].astype('str')
train['s_t'] = train['Source_Station_encode']+'_'+train['Target_Station_encode']
 
test['Source_Station_encode'] = test['Source_Station_encode'].astype('str') 
test['Target_Station_encode'] = test['Target_Station_encode'].astype('str')
test['s_t'] = test['Source_Station_encode']+'_'+test['Target_Station_encode']


x1 = train['s_t'].as_matrix().reshape([-1,1])
N1 = x1.shape[0]

x2 = test['s_t'].as_matrix().reshape([-1,1])
N2 = x2.shape[0]
x = np.concatenate((x1,x2))
x = preprocessing.LabelEncoder().fit_transform(x) #13963746x21567
print(x.shape)#(14454126, 21567)
train['new_s_t'] = x[:N1]
test['new_s_t'] = x[N1:]#0-14143
```

```python
from tqdm import tqdm, tqdm_pandas
tqdm.pandas()

def get_grid_history(s,t,w,p,r):
    b = hi[(hi['source_id']==s)&(hi['target_id']==t)&(hi['is_workday']==w)&(hi['is_peek']==p)&(hi['is_rain']==r)]
    if b.shape[0] > 0:
        return b.iloc[0]
    else :
        dic = {'Distance':[-1],'Diff_Time':[-1]}
        x = pd.DataFrame(dic)
        return x.iloc[0]
print('start')
tmp = trainx.progress_apply(lambda x : get_grid_history(int(x['Source_Station_encode']),int(x['Target_Station_encode']),x['is_workday'],x['is_peek'],x['is_rain']), axis=1)
trainx['encode_aver_diff'] = tmp['Diff_Time']
trainx['encode_aver_d'] = tmp['Distance']
trainx
```
