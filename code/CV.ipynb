```python
import numpy as np
import pandas as pd
import lightgbm as lgb
import pickle
from sklearn import metrics

with open('./data/train-id.txt', 'rb') as data_file:
    line = pickle.load(data_file)
col = [c for c in line if
       c not in ['O_LINENO', 'O_UP', 'Source_Station', 'Target_Station','O_TIME', 'aver_v','max_v', 'Diff_Time']]    
lgb_train  = lgb.Dataset(line[col].values, line['Diff_Time'].values)#feature_name=col,categorical_feature=['ID']


print('设置参数')
params = {
          'objective': 'regression_l2',
          'metric': 'rmse',
          'max_depth':8,
          'num_threads':20,
          }

### 交叉验证(调参)
print('交叉验证')
min_merror = float('Inf')
best_params = {}

# # 准确率
# print("调参1：提高准确率")
# for num_leaves in [100, 110, 120, 130, 140, 150]:
#     params['num_leaves'] = num_leaves

#     cv_results = lgb.cv(
#                         params,
#                         lgb_train,
#                         seed=2018,
#                         nfold=3,
#                         metrics=['rmse'],
#                         early_stopping_rounds=50,
#                         verbose_eval=True
#                         )

#     mean_merror = pd.Series(cv_results['rmse-mean']).min()
#     boost_rounds = pd.Series(cv_results['rmse-mean']).argmin()

#     if mean_merror < min_merror:
#         min_merror = mean_merror
#         best_params['num_leaves'] = num_leaves
            
# params['num_leaves'] = best_params['num_leaves']
# print(best_params['num_leaves'])
#100
params['num_leaves'] = 150
# 过拟合
print("调参2：降低过拟合")
for min_data_in_leaf in [10, 100,200, 300, 400, 500]:# 
    params['min_data_in_leaf'] = min_data_in_leaf

    cv_results = lgb.cv(
                        params,
                        lgb_train,
                        seed=2018,
                        nfold=3,
                        metrics=['rmse'],
                        early_stopping_rounds=50,
                        verbose_eval=True
                        )

    mean_merror = pd.Series(cv_results['rmse-mean']).min()
    boost_rounds = pd.Series(cv_results['rmse-mean']).argmin()

    if mean_merror < min_merror:
        min_merror = mean_merror
        best_params['min_data_in_leaf'] = min_data_in_leaf

params['min_data_in_leaf'] = best_params['min_data_in_leaf']
print(best_params['min_data_in_leaf'])
```

```python
print("调参3：降低过拟合")
for feature_fraction in [0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0]:
    params['feature_fraction'] = feature_fraction

    cv_results = lgb.cv(
                        params,
                        lgb_train,
                        seed=2018,
                        nfold=3,
                        metrics=['rmse'],
                        early_stopping_rounds=50,
                        verbose_eval=True
                        )

    mean_merror = pd.Series(cv_results['rmse-mean']).min()
    boost_rounds = pd.Series(cv_results['rmse-mean']).argmin()

    if mean_merror < min_merror:
        min_merror = mean_merror
        best_params['feature_fraction'] = feature_fraction

params['feature_fraction'] = best_params['feature_fraction']
print(best_params['feature_fraction'])

print("调参4：降低过拟合")
for lambda_l1 in [0.5, 1, 5, 10, 15, 20, 25]:
    for lambda_l2 in [0.5, 1, 5, 10, 15, 20, 25]:
        params['lambda_l1'] = lambda_l1
        params['lambda_l2'] = lambda_l2

        cv_results = lgb.cv(
                            params,
                            lgb_train,
                            seed=2018,
                            nfold=3,
                            metrics=['rmse'],
                            early_stopping_rounds=50,
                            verbose_eval=True
                            )

        mean_merror = pd.Series(cv_results['rmse-mean']).min()
        boost_rounds = pd.Series(cv_results['rmse-mean']).argmin()

        if mean_merror < min_merror:
            min_merror = mean_merror
            best_params['lambda_l1'] = lambda_l1
            best_params['lambda_l2'] = lambda_l2

params['lambda_l1'] = best_params['lambda_l1']
params['lambda_l2'] = best_params['lambda_l2']
print(best_params['lambda_l1'])
print(best_params['lambda_l2'])


print(best_params)
```
